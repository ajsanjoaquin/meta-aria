```python
from langchain.chat_models import ChatOpenAI
from langchain.chains.summarize import load_summarize_chain
from langchain.docstore.document import Document
from langchain.document_loaders.csv_loader import CSVLoader
from langchain import PromptTemplate

prompt_template = """ Summarize CSV file about user's travel taste. Make a lot of "user has a preference for" statements to find connection with other users' interests. Keep summary under 5 to 10 sentences.
Timeline:

{text}


TL'DR: """

os.environ["OPENAI_API_KEY"] = ""
PROMPT = PromptTemplate(template=prompt_template, input_variables=["text"])
llm = ChatOpenAI(temperature=0, model_name="gpt-3.5-turbo")
chain = load_summarize_chain(llm, chain_type="stuff", prompt=PROMPT)
#CHANGE PATH
loader = CSVLoader('/content/personA_map.csv')

docs = loader.load()

chain.run(docs)
```
